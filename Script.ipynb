```python
import pandas as pd
import os
```

```python
folder_path = "March"
all_data =[]
```

```python
columns_to_keep = [
    'Time', 'EacToday(kWh)', 'EacTotal(kWh)', 'Ppv(W)', 'EpvToday(kWh)', 'EpvTotal(kWh)', 'TimeTotal(S)',
    'IacR(A)', 'IacS(A)', 'IacT(A)', 'PacR(VA)', 'PacS(VA)', 'PacT(VA)', 
    'VacR(V)', 'VacS(V)', 'VacT(V)', 'Pac(W)', 'INVTemp(℃)', 'AMTemp1(℃)', 
    'BTTemp(℃)', 'OUTTemp(℃)', 'AMTemp2(℃)', 'ReactPower(Var)', 
    'ReactPowerMax(Var)', 'ReactPower_Total(kWh)'
]
```

```python
for filename in os.listdir(folder_path):
    if filename.endswith('.xls'):
        file_path = os.path.join(folder_path, filename)
        try:
            # Skip first 3 metadata rows
            df = pd.read_excel(file_path, skiprows=3)

            # Keep only the needed columns
            df = df[columns_to_keep]

            all_data.append(df)
        except Exception as e:
            print(f"❌ Error processing {filename}: {e}")
```

```python
print(type(all_data))
```

```python
combined_df = pd.concat(all_data, ignore_index=True)
```

```python
combined_df
```

```python
# Parse 'Time' column to datetime
combined_df['Time'] = pd.to_datetime(combined_df['Time'], errors='coerce')
```

```python
combined_df
```

```python
combined_df = combined_df.dropna(subset=['Time'])
combined_df.shape
```

```python
combined_df.to_csv('march_cleaned.csv', index=False)
```
